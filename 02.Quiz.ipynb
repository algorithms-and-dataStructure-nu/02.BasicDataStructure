## 課題: Assignment
リスト`A`の各要素に対して、ある演算を行って新しいリスト`B`を作成するプログラムを作成することを考える。
もちろん、`for`文を用いて、リスト`A`の各要素に対して演算を行い、その結果をリスト`B`に追加することができる。
しかし、Pythonにはリスト内包表記という便利な機能がある。
この機能を用いて、`A=[1, 2, 3, 4, 5]`の各要素を2倍して新しいリスト`B`を作成するプログラムを以下に示す。
リスト内包表記を用いることで、`A`の各要素を2乗して新しいリスト`B`を簡潔に作成するプログラムを作成し、動作を確認しなさい。
なお、`set`や`dict`などの他のデータ構造に対しても、同様の方法で内包表記を用いることができることに留意しなさい。

Let us consider creating a program that performs a certain operation on each element of a list `A` to create a new list `B`. Of course, we can use a `for` loop to perform the operation on each element of list `A` and append the result to list `B`. However, Python has a convenient feature called list comprehension. 
The following program demonstrates how to use list comprehension to create a new list `B` by doubling each element of `A=[1, 2, 3, 4, 5]`.
Use this feature to create a program that squares each element of `A=[1, 2, 3, 4, 5]` to create a new list `B`, and verify its operation.
For other data structures such as `set` and `dict`, similar methods can be used with comprehensions, so keep that in mind.

```python
A = [1, 2, 3, 4, 5]
B = [x * 2 for x in A]
print(B)  # Output: [2, 4, 6, 8, 10]
```
